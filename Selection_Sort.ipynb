```python
# Selection Sort algorithm for double array arr according to column with index col
def selection_sort(arr, col):
  n = len(arr)
  for i in range(1,n):
    min_index = i
    for j in range(i + 1, n):
      if float(arr[j][col]) < float(arr[min_index][col]):
        min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
  return arr
```

```python
import csv
file = csv.reader(open('Gender Wage Gap.csv', 'r'))
rows = [row for row in file]

sorted_rows = [', '.join(row) + '\n' for row in selection_sort(rows,6)]
with open('output.csv', 'w+') as f:
    f.writelines(sorted_rows)
pd.read_csv("output.csv")
```
